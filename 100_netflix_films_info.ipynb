```python
film_review_div = soup.find('div',class_='articleContentBody')

for po in film_review_div:
    print(po)
```

```python
films_full_info = []
def beast_movies_on_netflix(index_no):
    try:

        film_info = film_review_div.find(id='row-index-'+str(index_no))
        title = film_info.find('h2').getText()
        film_body = film_info.find('div',class_='info critics-consensus')
        film_body = film_body.getText().replace('Critics Consensus:','')

        film_synopsis = film_info.find('div',class_='info synopsis')
        Synopsis_text = film_synopsis.getText().replace('Synopsis:','')
        film_synopsis_span = film_synopsis.find('span').getText().replace(':','')

        dire_class = film_info.find('div',class_='info cast')
        director = dire_class.getText().replace('Starring:','')

        span_class = dire_class.find('span',class_='descriptor').getText().replace(':','')

        links = dire_class.find_all('a')

        directors = []
        for text in links:
            directors.append(text.getText())

        directors_div = film_info.find('div',class_='info director')
        directors_link = dire_class.find('a').getText()

        
        directors_span = directors_div.find('span').getText().replace(':','')

        films = {'title ': title,
                'Critics Consensus ': film_body,
                film_synopsis_span : Synopsis_text,
                span_class : directors,
                directors_span :directors_link

                
                }

        films_full_info.append(films)
    except Exception as e :
        print(e)

    return films_full_info

```

```python
for page_no in range(1,100):
    beast_movies_on_netflix(page_no)
```

```python
def save_data(title, data):
    with open(title,'w',encoding='utf-8') as f:
        json.dump(data, f, ensure_ascii=False,indent=4)

#it loads the data
def load_data(title):
    with open(title,encode='utf-8') as f:
        return json.load(f)
```
